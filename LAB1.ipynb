#EXERCISE 3

```python
def reverse(xlist):
    reversed_list = []
    index = len(xlist) - 1
    while index >= 0:
          reversed_list.append(xlist[index])
          index -= 1
    return reversed_list

print(reverse([10,20,30,40,50]))
```
